```python
import pandas as pd
data = {'model': ['   ULMFit-SSP1', 'RF', 'GB', 'XGBOOST', 'DNN', 'SVM'],
        'Test AUC-ROC': [0.8416, 0.9468, 0.9410, 0.9285, 0.9278, 0.9356],
        'F1 score': [0.8117, 0.9372, 0.9324, 0.9217, 0.9308, 0.9251],
        'Test accuracy': [0.8247, 0.9005, 0.8939, 0.8978, 0.9025, 0.9052]}

df = pd.DataFrame(data)
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def get_unique_sorted_values(start, end, step=0.01, num_decimals=4):
    return sorted({round(x, num_decimals) for x in np.arange(start - step, end + step, step)})



# Set the style of the plot
sns.set(style="ticks")

# Create a figure and axis
fig, ax = plt.subplots(figsize=(5, 4), subplot_kw=dict(polar=True, facecolor='whitesmoke'))  # Set the facecolor

# Get the number of models
num_models = len(df)

# Calculate angle and axis positions for the radar plot
angles = [n / float(num_models) * 2 * np.pi for n in range(num_models)]
angles += angles[:1]

# Plot each model's F1 score with markers and labels
for i, model in enumerate(df['model']):
    values = df.loc[i, 'Test AUC-ROC']
    print(values)
    ax.plot(angles[i], float(values), marker='o', markersize=8, label=model, color=plt.get_cmap('Set1')(i/num_models))
    # Add label annotations with black color
    #ax.annotate(f'{float(values):.4f}', xy=(angles[i], float(values)), xytext=(5, 5), textcoords='offset points', color='black')

# Create a polygon to fill the area between connected points with light blue color
polygon_data = []
for i in range(num_models):
    values = df.loc[i, 'Test AUC-ROC']
    polygon_data.append([angles[i], float(values)])
    print(polygon_data)
polygon = plt.Polygon(polygon_data, closed=True, color='purple', alpha=0.5)
ax.add_patch(polygon)

# Customize y-axis limit
ax.set_ylim([0.6500, 0.9700])

# Set the labels for each axis with black color and bold
ax.set_xticks(angles[:-1])
ax.set_xticklabels(df['model'], color='black')

ax.set_yticklabels([])

# Add grid lines with light black color
ax.grid(True, linestyle='dashed', linewidth=0.5, color='dimgray', alpha=0.7)

# Add 'AUC-ROC' label outside the radar plot
# Add 'AUC-ROC' label outside the radar plot on the left side
# Add 'AUC-ROC' label on top of the radar plot
plt.figtext(0.54, 1.03, 'AUC score', rotation='horizontal', ha='center', va='center', fontsize=12, color='black')



# Save the plot as PNG
plt.savefig('/raid/aiccg/rbsunoj/Nupur/AUC-new.png', bbox_inches='tight', dpi=300)

# Show the plot
plt.show()
```
