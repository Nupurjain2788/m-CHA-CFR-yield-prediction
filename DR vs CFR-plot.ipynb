```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
df_sorted = pd.read_csv("/raid/aiccg/rbsunoj/supratim/sorted_5percent_predictions.csv")

# Creating subplots for each column
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(12, 8))

# Plotting direct_regression
axes[0].bar(range(len(df_sorted)), df_sorted['direct_regression'], color='red', label='direct regression')
axes[0].legend()


# Plotting CFR
axes[1].bar(range(len(df_sorted)), df_sorted['CFR'], color='blue', label='CFR')
axes[1].legend()
axes[1].set_xlabel('reaction')

axes[1].set_ylim(0, 50)

plt.tight_layout()
plt.savefig('/raid/aiccg/rbsunoj/Nupur/predictions_metach/direct_and_CFR_plot.png', dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
df_sorted = pd.read_csv("/raid/aiccg/rbsunoj/supratim/sorted_5percent_predictions.csv")

# Create a single figure with one subplot
fig, ax = plt.subplots(figsize=(12, 6))

# Define the width of the bars and create an offset for plotting them side by side
bar_width = 0.4
index = range(len(df_sorted))

# Plotting direct_regression with a slight shift to the left
ax.bar([i - bar_width/2 for i in index], df_sorted['direct_regression'], width=bar_width, color='#FF6666', label='DR')

# Plotting CFR with a slight shift to the right
ax.bar([i + bar_width/2 for i in index], df_sorted['CFR'], width=bar_width, color='blue', label='CFR')

# Adding legend
ax.legend()

# Labeling the axes
ax.set_xlabel('Reaction', fontsize=14)
ax.set_ylabel('Values', fontsize=14)

# Increasing the font size of the tick labels on both x and y axes
ax.tick_params(axis='both', which='major', labelsize=10)

# Set limits for y-axis if needed
ax.set_ylim(0, 57)

# Apply tight layout
plt.tight_layout()

plt.savefig('/raid/aiccg/rbsunoj/Nupur/predictions_metach/direct_and_CFR_plot-combine-2.png', dpi=300)
plt.show()
```
